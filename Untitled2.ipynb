```python
distancia = float(input("Ingresa la distancia recorrida (en metros): "))
tiempo = float(input("Ingresa el tiempo (en segundos): "))

if tiempo > 0:
    velocidad = distancia / tiempo
    print("La velocidad es:", velocidad, "m/s")
else:
    print("Error: el tiempo no puede ser cero.")
```

```python
import math


num1 = int(input("Ingresa el primer número: "))
num2 = int(input("Ingresa el segundo número: "))

mcd = math.gcd(num1, num2)
mcm = abs(num1 * num2) // mcd

print(f"El MCM de {num1} y {num2} es: {mcm}")

```
